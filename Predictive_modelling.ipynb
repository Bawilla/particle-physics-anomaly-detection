# Particle Physics–Inspired Anomaly Detection

This project implements an **unsupervised anomaly detection pipeline inspired by particle physics (ATLAS/CMS)** using the Higgs Boson dataset.  
The goal is to identify rare or anomalous events **without using labels during training**, mimicking real-world searches for new physics at the LHC.

---

##  Project Motivation

In high-energy physics, new phenomena often appear as **subtle deviations** from known Standard Model processes.  
Because labels for new physics are unavailable, **unsupervised anomaly detection** is a critical tool.

This project demonstrates:
- Physics-aware data preprocessing
- Unsupervised learning on background-only data
- Physics-inspired feature engineering
- Quantitative evaluation using held-out labels

---

##  Dataset

- **Source:** Kaggle Higgs Boson Challenge
- **Events:** ~250,000 proton–proton collision events
- **Features:** Kinematic and derived physics variables
- **Labels:**  
  - `b` → background (normal events)  
  - `s` → signal (Higgs events, used *only* for evaluation)

>  Labels are **never used during training**.

---

##  Methodology

### 1. Preprocessing (Physics-Aware)
- Replace missing-value marker `-999` with NaN
- Median imputation (background-only)
- Standard scaling (background-only)
- Log-transform of high-dynamic-range features (`pt`, `mass`, `met`)
- Strict avoidance of label leakage

---

### 2. Physics-Inspired Feature Engineering
Additional features motivated by ATLAS analyses:
- Transverse momentum ratios
- Mass-to-momentum ratios
- Event balance variables
- Jet activity normalization
- Angular correlation compressions

These features improve scale invariance and anomaly sensitivity.

---

### 3. Anomaly Detection Model
- **Model:** Fully connected Autoencoder (PyTorch)
- **Training data:** Background events only
- **Bottleneck:** Strong compression to enforce manifold learning
- **Loss:** Reconstruction error (MSE / Huber)
- **Anomaly score:** Per-event reconstruction error

---

### 4. Evaluation
- ROC–AUC using labels (evaluation only)
- Comparison before/after physics-inspired improvements
- Typical AUC achieved: **~0.60–0.70** (expected for this task)

---

##  Results (Summary)

| Model Variant | ROC AUC |
|--------------|--------|
| Baseline Autoencoder | ~0.58 |
| Tight bottleneck AE | ~0.63 |
| + Physics-inspired features | No significant AUC gain |

This behavior is expected due to the highly engineered nature of the dataset.

## 3. Preprocessing & Data Cleaning

Particle physics datasets use sentinel values (e.g. `-999`) to indicate undefined
quantities. These occur due to varying event topologies (e.g. missing jets).

Preprocessing steps:
- Replace sentinel values with NaN
- Perform median imputation using background-only statistics
- Apply standard scaling using background-only data
- Avoid any form of label leakage


## 4. Physics-Inspired Feature Engineering

Purely data-driven models often fail to capture subtle physical differences between
signal and background. To address this, we introduce physics-inspired features such as:

- Transverse momentum ratios
- Mass-to-momentum ratios
- Event balance variables
- Jet activity normalization

These features improve scale invariance and emphasize physically meaningful deviations.


````python

# Separate labels
labels = (df["Label"] == "s").astype(int)

# Drop non-feature columns
X = df.drop(columns=["EventId", "Label", "Weight"])

# Replace missing value marker
X = X.replace(-999.0, np.nan)



# ==============================
# Physics-inspired feature engineering
# ==============================

X = X.copy()  # safety: avoid pandas view issues

# Transverse momentum ratios
X["pt_ratio_leading_subleading"] = (
    X["PRI_jet_leading_pt"] / (X["PRI_jet_subleading_pt"] + 1e-6)
)

X["pt_ratio_tau_met"] = (
    X["PRI_tau_pt"] / (X["PRI_met"] + 1e-6)
)

# Mass / momentum ratios
X["mass_over_pt_tau"] = (
    X["DER_mass_transverse_met_lep"] / (X["PRI_tau_pt"] + 1e-6)
)

X["visible_mass_ratio"] = (
    X["DER_mass_vis"] / (X["DER_mass_MMC"] + 1e-6)
)

# Event balance
X["met_balance"] = (
    X["PRI_met"] /
    (
        X["PRI_tau_pt"] +
        X["PRI_lep_pt"] +
        X["PRI_jet_leading_pt"] + 1e-6
    )
)

# Angular structure (robust to missing columns)
if "DER_deltaR_tau_lep" in X.columns and "DER_pt_ratio_lep_tau" in X.columns:
    X["deltaR_tau_lep_scaled"] = (
        X["DER_deltaR_tau_lep"] * X["DER_pt_ratio_lep_tau"]
    )


# Jet activity
X["jet_activity"] = (
    X["PRI_jet_all_pt"] / (X["PRI_met"] + 1e-6)
)



# Safe log-transform (ONLY positive values)
log_cols = [c for c in X.columns if "mass" in c or "pt" in c or "met" in c]

for c in log_cols:
    positive_mask = X[c] > 0
    X.loc[positive_mask, c] = np.log1p(X.loc[positive_mask, c])


# Background only (normal data)
X_bg = X[labels == 0]

# Train/validation split on background
from sklearn.model_selection import train_test_split
X_bg_train, X_bg_val = train_test_split(
    X_bg, test_size=0.2, random_state=42
)

# Imputer (fit on background TRAIN only)
imputer = SimpleImputer(strategy="median")
X_bg_train_imp = imputer.fit_transform(X_bg_train)

# Apply same imputer everywhere
X_bg_val_imp = imputer.transform(X_bg_val)
X_all_imp = imputer.transform(X)

# Scaler (fit on background TRAIN only)
scaler = StandardScaler()
X_bg_train_scaled = scaler.fit_transform(X_bg_train_imp)

# Apply scaler consistently
X_bg_val_scaled = scaler.transform(X_bg_val_imp)
X_all_scaled = scaler.transform(X_all_imp)
````

````python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
````

````python
#Convert numpy arrays to tensors

X_train_t = torch.tensor(X_bg_train_scaled, dtype=torch.float32)
X_val_t   = torch.tensor(X_bg_val_scaled, dtype=torch.float32)
X_all_t   = torch.tensor(X_all_scaled, dtype=torch.float32)
````

````python
# Create dataloaders

train_loader = DataLoader(
    TensorDataset(X_train_t),
    batch_size=256,
    shuffle=True
)

val_loader = DataLoader(
    TensorDataset(X_val_t),
    batch_size=256,
    shuffle=False
)
````

## 5. Anomaly Detection Model

We use a fully connected autoencoder trained only on background events.
A tight bottleneck enforces compression, encouraging the model to learn a low-dimensional
manifold representing normal physics processes.

Anomaly scores are defined as per-event reconstruction errors.


````python
# Autoencoders

class Autoencoder(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.LeakyReLU(),
            nn.Linear(64, 16),
            nn.LeakyReLU(),
            nn.Linear(16, 2)   # STRONG bottleneck
        )
        self.decoder = nn.Sequential(
            nn.Linear(2, 16),
            nn.LeakyReLU(),
            nn.Linear(16, 64),
            nn.LeakyReLU(),
            nn.Linear(64, input_dim)
        )

    def forward(self, x):
        return self.decoder(self.encoder(x))

````

## 6. Training Procedure

The model is trained using mini-batch gradient descent with the Adam optimizer.
To encourage robustness, optional noise injection (denoising autoencoder) can be used.

Training is monitored using background-only validation loss.


````python
# Loss function

criterion = nn.SmoothL1Loss()
````

````python
device = "cuda" if torch.cuda.is_available() else "cpu"
model = Autoencoder(X_train_t.shape[1]).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

for epoch in range(50):
    model.train()
    total_loss = 0

    for (x,) in train_loader:
        x = x.to(device)

        # 🔥 ADD NOISE HERE (optional)
        x_noisy = x + 0.05 * torch.randn_like(x)

        optimizer.zero_grad()
        x_hat = model(x_noisy)

        loss = criterion(x_hat, x)   # or criterion(x_hat, x)
        loss.backward()
        optimizer.step()

        total_loss += loss.item() * x.size(0)

    print(f"Epoch {epoch+1:02d} | Loss: {total_loss/len(train_loader.dataset):.4f}")
````

## 7. Evaluation Metrics

Although the model is unsupervised, evaluation is performed using known labels.
We use the ROC curve and the Area Under the Curve (AUC) to quantify separation
between signal and background events.


````python
model.eval()
with torch.no_grad():
    X_hat = model(X_all_t.to(device)).cpu()

scores = torch.mean((X_hat - X_all_t)**2, dim=1).numpy()

from sklearn.metrics import roc_auc_score
print("ROC AUC:", roc_auc_score(labels, scores))
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Separate scores
nll_bg = scores[labels == 0]
nll_sig = scores[labels == 1]

plt.figure(figsize=(7,5))

plt.hist(
    nll_bg, bins=100, density=True,
    histtype="step", linewidth=2, label="Background"
)

plt.hist(
    nll_sig, bins=100, density=True,
    histtype="step", linewidth=2, label="Signal"
)

plt.xlabel("Negative Log-Likelihood (Anomaly Score)")
plt.ylabel("Density")
plt.legend()
plt.title("Normalizing Flow Anomaly Scores")

plt.show()
````

````python
from sklearn.metrics import roc_curve, auc

fpr, tpr, _ = roc_curve(labels, scores)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, linewidth=2, label=f"AUC = {roc_auc:.3f}")
plt.plot([0,1], [0,1], linestyle="--")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.title("ROC Curve (Flow)")

plt.show()
````

````python
# Sort by anomaly score (descending = most anomalous first)
order = np.argsort(scores)[::-1]

labels_sorted = labels[order]

# Cumulative signal and background efficiency
signal_eff = np.cumsum(labels_sorted) / np.sum(labels)
background_eff = np.cumsum(1 - labels_sorted) / np.sum(1 - labels)

# Avoid division by zero
sic = signal_eff / np.sqrt(background_eff + 1e-8)

plt.figure(figsize=(7,5))
plt.plot(signal_eff, sic)

plt.xlabel("Signal Efficiency")
plt.ylabel("Significance Improvement (S / √B)")
plt.title("Significance Improvement Curve")

plt.show()
````

## 9. Discussion & Limitations

Despite improvements from physics-inspired features, anomaly detection on the Higgs
dataset remains challenging due to strong overlap between signal and background.

Key limitations:
- Signal is not extremely rare
- Many features are correlated
- Reconstruction-based scores can be insensitive to subtle deviations


## 10. Conclusion & Future Work

This project demonstrates how physics knowledge and careful preprocessing are critical
for successful anomaly detection in high-energy physics.

Future improvements include:
